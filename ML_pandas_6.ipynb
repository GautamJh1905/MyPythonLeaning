```python
import pandas as pd
df = pd.read_csv("Emp_Data.csv")
df1 = pd.read_csv("department.csv")
print(df)
print(df1)
```

```python
# Reload df to ensure clean state
df = pd.read_csv("Emp_Data.csv")

# Map department names to department_id in df
dept_mapping = dict(zip(df1["department"], df1["department_id"]))
df["department_id"] = df["department"].map(dept_mapping)

# Drop the old department column
df = df.drop("department", axis=1)

# Merge df with df1 on department_id
merged_df = df.merge(df1, on="department_id", how="left")

# Save merged data to new CSV file
merged_df.to_csv("merged.csv", index=False)

print("Merged data saved to merged.csv")
print("\nMerged DataFrame:")
print(merged_df)

# Show employee count per department
print("\nEmployee Count per Department:")
employee_count = merged_df.groupby("department")["emp_id"].count()
print(employee_count)
```
