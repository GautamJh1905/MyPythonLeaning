```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("Emp_Data.csv")

# Count occurrences of each age
age_counts = df["age"].value_counts()

# Create pie chart with count instead of percentage
plt.figure(figsize=(8, 8))
plt.pie(age_counts.values, labels=age_counts.index, autopct=lambda pct: f'{int(pct/100.*sum(age_counts.values))}')
plt.title("Age Distribution - Employee Count")
plt.show()
```
